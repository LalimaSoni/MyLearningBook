##### Import required python libraries

### Reading Restaurant_data.json file and loading into FileDf Dataframe.

```python
FileDf = pd.read_json("https://raw.githubusercontent.com/ashraf356/cc4braininterview/main/restaurant_data.json")
FileDf.head()
```

#### Number of rows and columns in FileDf is 479, 7 respectively

```python
FileDf.shape
```

###### We can observe that in FileDf , column 'restaurants' has list of elements. To  transform each element of a list-like to a row, we use here explode( ) function.

```python
FileDf = FileDf.explode('restaurants')
FileDf.shape
```

##### After explode() number of rows increases to 1600, columns are still 7.

```python
FileDf.head()
```

##### Now creat new Dataframe 'RestaurantsDf' by converting FileDf to json then load and then normalize to get different columns from list like rows. 

```python
RestaurantsDf = pd.json_normalize(json.loads(FileDf.to_json(orient = "records")))
RestaurantsDf.shape
```

#####  'RestaurantsDf' dataframe has 1600 rows and 45 columns

```python
RestaurantsDf.head()
```

##### similar as above,  'RestaurantsDf' dataframe's column 'restaurants.restaurant.zomato_events'  has list of elements.Repeating above process to get normalize form.

```python
RestaurantsDf = RestaurantsDf.explode('restaurants.restaurant.zomato_events')
RestaurantsDf.shape
```

```python
RestaurantsDf.head()
```

```python
EventDf = pd.json_normalize(json.loads(RestaurantsDf.to_json(orient = "records")))
EventDf.shape
```

```python
EventDf.head()
```

```python
EventDf = EventDf.explode('restaurants.restaurant.zomato_events.event.photos')
EventDf.shape
```

```python
PhotoDf = pd.json_normalize(json.loads(EventDf.to_json(orient = "records")))
PhotoDf.shape
```

```python
PhotoDf.head()
```

### Solution4 (Optional ETL challenge:)
○ From the dataset, use data analytics to determine the threshold for the different <b>“rating_text” </b>based on <b>“aggregate_rating”

```python
Solution4 = Sol_merge_country[['Restaurant_id','User_rating_text','User_agg_rating']]
Solution4 = Solution4.drop_duplicates()
Solution4 = Solution4[['User_rating_text','User_agg_rating']]
Solution4.head()
```

```python
Solution4.shape
```

##### Considering Median of 'user_agg_ rating' for threshold

```python
Solution4.groupby('User_rating_text').median()
```

##### Calculating 'Rating_median' and 'outlier' for every rating

```python
Solution4['Rating_median'] = Solution4.groupby('User_rating_text').transform('median')
difference = (Solution4.User_agg_rating - Solution4['Rating_median']).abs()
Solution4['outlier'] = ( difference > 0.5 ) 
Solution4.head()
```

##### There are two outlier for Average rating.
As we can observe below  user_agg_rating is 2.6 and 2.5, These two ratings should belongs to "Poor", but given under "Average"

```python
print( Solution4[(Solution4.outlier  == True)] )
```
